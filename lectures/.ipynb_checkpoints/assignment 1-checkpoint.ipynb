# provide the code, explanation, and scatter plot in an ipynb file such that we can run it on Jupyter notebook. 


```python
# generate N Gaussian samples  and compute the sample mean
import numpy as np
from matplotlib import pyplot
mean = [10, 20]
cov = [[3, 4], [4, 8]]
N = 5000
sample = np.random.multivariate_normal(mean, cov, N)
sample
x,y = sample.T
pyplot.scatter(x,y, color = 'rgb')
# show line plot
pyplot.show()
```

```python

x_bar= np.sum(x, axis=0)/N
y_bar = np.sum(y, axis= 0)/N
x_center, y_center = (x-x_bar),(y-y_bar) 
### this is the plot after centering!!!!
pyplot.scatter(x_center,y_center, color = 'rgb')
# show line plot
pyplot.show()
```

```python
sample_cov = np.cov(x,y)
sample_cov
```

```python
Ev,E = np.linalg.eig(sample_cov)
print(Ev)
print(E)
```

```python
xt = np.inner(E,sample)
```

```python
xtx, xty = xt
pyplot.scatter(xtx,xty, color = 'rgb')
# show line plot
pyplot.show()
```

based on the scatteplot, we can see that the data have been shifted to the left. Also, it is more concentrated towards the center which shows improvement in correlation.

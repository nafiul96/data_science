1. Determine the average number of patients for each patient concondition and sex
2. Determine the number of patients per patient condition and insurance type
3. Create a new column and that shows the visit time of each patient. Name this column 'timeWithDr'
4. Determine the day and month of visit for each patient
5. Fit a linear regression model to the data to investigate if you can predict the time a  doctor spends with a  patient from the rest of data. Report the accuracy of your regression model by r2-squared and mean squared error. Also plot the actual and estimated time difference.

6. Using a Random forest classifier  predicts the patient condition given the other set of features. Report the results in a confusion matrix


```python
# read the excel file
import pandas as pd
import numpy as np
# The excel file path
report_path = '/users/nafi/downloads/CSE391MidtermData.xlsx'

xls = pd.ExcelFile(report_path)
data = pd.read_excel(xls) 
data
```

```python
# print the average of patients for each patient con
# condition and sex

table = pd.pivot_table(data,columns=["PATIENT_CONDITION","INSURANCE_TYPE"],aggfunc=[np.sum])
table
```

```python
# print the number of patients per patient condition and insurance type

table = pd.pivot_table(data,columns=["PATIENT_CONDITION","SEX"],aggfunc=[np.sum])
table

```

```python
import datetime  as datatime 
# determine the time that a doctor spends with a patient

# add this time to the data frame
time_difference = data['PT_END_TIME'] - data['PT_START_TIME']
data['timeWithDr'] = time_difference


# determine the day and month of visit for each patient
visitday= data['VISIT_DATE'].map(lambda x: datatime.datetime.strptime(str(x),'%Y-%m-%d %H:%M:%S'))
data['day'] = [int(i.day) for i in visitday ]
data['month'] = [int(i.month) for i in visitday ]

#data
#data.drop(columns=['VISIT_DATE','PT_START_TIME', 'PT_END_TIME'])
data= data.drop(['VISIT_DATE','PT_START_TIME', 'PT_END_TIME'], axis =1)
#data.drop([5,6,7])
data
```

```python

# fit the regression model to the data to investigate if
# we can predict the time a given doctor spends for a given patient
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_predict

from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score




# PATIENT_CONDITION	INSURANCE_TYPE	AGE	SEX	PROVIDER_NAME	timeWithDr	day	month

# doctor name -> time with doctor
data["PATIENT_CONDITION"] = pd.get_dummies(data["PATIENT_CONDITION"])
data["INSURANCE_TYPE"] = pd.get_dummies(data["INSURANCE_TYPE"])
data["SEX"] = pd.get_dummies(data["SEX"])
data["PROVIDER_NAME"] = pd.get_dummies(data["PROVIDER_NAME"])


#  PATIENT_CONDITION	INSURANCE_TYPE	AGE	SEX	PROVIDER_NAME	timeWithDr	day	month

X = data.ix["PATIENT_CONDITION","INSURANCE_TYPE","AGE", "SEX","PROVIDER_NAME","day", "month"]
y = data["timeWithDr"]

reg = LinearRegression().fit(X, y)
reg.score(X, y)
 
















```

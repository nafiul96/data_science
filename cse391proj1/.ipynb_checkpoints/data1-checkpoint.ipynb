```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd
import scipy.io as sciIO
import scipy.cluster.vq as vq
import matplotlib.pyplot as plt
import numpy as np
import re

# part 1
df = pd.read_csv("/Users/katchoi/Projects/Cancer_Data.txt", sep='\t')
df = df.set_index('gene')
df = df[16:]

# 1 - randomly 2,000 genes out of 18,026 and 50 patients out of 500 patients in this dataset
df = df.fillna(df.median(numeric_only = True, skipna = True, axis = 0))

sample = df.sample(n=2000).sample(n=50, axis=1)
centroid , label = vq.kmeans2(sample, 20, minit='random')
sample = sample.as_matrix()
plt.rcParams.update({'font.size': 12})

for c in range(0,20):
    plt.subplot(4,5,c+1)
    p = np.where(label==c)
    plt.plot(list(range(0,50)),sample[p].T)
plt.title('K-Means Clustering of Profiles')
plt.show()
# print(np.array(np.shape(label)))
# print(np.array(np.shape(centroid)))

# Plot the centers of each cluster and  
for c in range(0,20):
    plt.subplot(4,5,c+1)
    plt.plot(list(range(0,50)),centroid[c,:])  
    plt.axis('off')
plt.title('K-Means Centroid')
plt.show()

print("\nthe number of genes in each cluster:")
print(centroid[0].size)

# Do you think you need to increase or decrease the number of cluster in order to have a 
# better clustering for these data. - dk

plt.rcParams.update({'font.size': 24})
# hierarchical clustering on patients and display the dendrogram
corrDist = ssp.pdist(sample.T, 'correlation')
clusterTree = hi.linkage(corrDist, method='average')
clusters = hi.fcluster(clusterTree, 16, criterion= 'maxclust')

# calculate full dendrogram
plt.figure(figsize=(25, 10))
dn = hi.dendrogram(clusterTree)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('patients')
plt.ylabel('distance')
plt.show()

# print(np.array(np.shape(sample)))
# print(np.array(np.shape(clusterTree)))

# 5 - perfrom pca analysis on patience
pca = PCA(n_components=5)
pca.fit(sample.T)

# Compute how much of variation is explained by each principle component
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.size)
print(pca.explained_variance_)
print(pca.explained_variance_.size)

print("\nIf you want to compress these data, how many PCs you would choose:")
print("Less number of components")
```
